### Algerian Forest Fire Dataset

```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv('Algerian_forest_fires_dataset.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df[df.isnull().any(axis=1)]
```

```python
df.loc[:122,'Region']=0
df.loc[122:,'Region']=1
```

```python
df['Region']=df['Region'].astype(int)
```

```python
df.head()
```

```python
df=df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.iloc[122]
```

```python
df=df.drop(122).reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.columns=df.columns.str.strip()
```

```python
df.columns
```

```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']]=df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df[['Rain', 'FFMC','DMC', 'DC', 'ISI', 'BUI', 'FWI']]=df[['Rain', 'FFMC','DMC', 'DC', 'ISI', 'BUI', 'FWI']].astype(float)
```

```python
df.info()
```

```python
df.to_csv('clean_dataset1',index=False)
```

```python
percentage=df['Classes'].value_counts(normalize=True)*100
```

```python
classlabes=['fire','not fire']  # Assuming these are the desired labels
percentage = [70, 30, 5]  # Assuming this has extra data

# Truncate percentage if it has more elements than labels
if len(percentage) > len(classlabes):
    percentage = percentage[:len(classlabes)]  # Keep only the first two elements

plt.figure(figsize=(12,7))
plt.pie(percentage, labels=classlabes, autopct='%1.1f%%')
plt.title("Pie Chart of Classes")
plt.show()
```
